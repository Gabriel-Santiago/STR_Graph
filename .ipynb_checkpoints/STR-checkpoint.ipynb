```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura dos txts
dados = pd.read_fwf('dados.txt')
dados2 = pd.read_fwf('dados2.txt')
dados3 = pd.read_fwf('dados3.txt')
```

```python
# Maior valor do dados
piorTempoDeRespostaTemp = dados.max()
piorTempoDeRespostaTemp
```

```python
# Maior valor do dados
piorTempoDeRespostaNivel = dados2.max()
piorTempoDeRespostaNivel
```

```python
indice = np.arange(999)
```

```python
# Adicionando indice ao txt de dados
data = pd.DataFrame(dados)
data['Índice'] = indice
data
```

```python
# Uma figura onde no eixo Y tem-se o tempo de resposta do controlador de temperatura e no 
# eixo X tem-se o índice de uma dada amostra que foi salva.

fig, ax = plt.subplots()

ax.scatter(data['Índice'], data['3756'])
ax.set_xlabel("Quantidade de amostras")
ax.set_ylabel("Tempos de resposta")
ax.set_title('Gráfico do tempo de reposta do controlador de temperatura')

plt.show()
```

```python
# Adicionando indice ao txt de dados2
data2 = pd.DataFrame(dados2)
data2['Índice'] = indice
data2
```

```python
# Uma figura onde no eixo Y tem-se o tempo de resposta do controlador de nível de água e no 
# eixo X tem-se o índice de uma dada amostra que foi salva.

fig, ax = plt.subplots()

ax.scatter(data2['Índice'], data2['3732'])
ax.set_xlabel("Quantidade de amostras")
ax.set_ylabel("Tempos de resposta")
ax.set_title('Gráfico do tempo de reposta do controlador de nível de água')

plt.show()
```

```python
# Histograma dos tempos de resposta do controlador da temperatura
fig, ax = plt.subplots()

ax.hist(data['3756'])
ax.set_xlabel("Tempos de resposta")
ax.set_ylabel("Quantidade de amostras")
ax.set_title('Gráfico do tempo de reposta do controlador de temperatura')

plt.show()
```

```python
# Histograma dos tempos de resposta do controlador do nível de água
fig, ax = plt.subplots()

ax.hist(data2['3732'])
ax.set_xlabel("Tempos de resposta")
ax.set_ylabel("Quantidade de amostras")
ax.set_title('Gráfico do tempo de reposta do controlador do nível de água')

plt.show()
```

```python
# Tratar dados 3 para separar os dados de temperatura e nível
# Vou chamar o dados3 para verificar a temperatura e nível
dados3
```

```python
# O nível está nos índices pares e a temperatura nos ínidces ímpares
# Vou adicionar uma coluna para pegar os índices
data3 = pd.DataFrame(dados3)
data3['Índice'] = indice
data3
```

```python
# data4 será para a temperatura de água
data4 = data3[data3['Índice']%2 != 0]
data4
```

```python
# Figura contendo evolução temporal da temperatura da água
data4['28.007041'].plot()
```

```python
# data5 será para o nível de água
data5 = data3[data3['Índice']%2 == 0]
data5
```

```python
# Figura contendo evolução temporal do nível da água
data5['28.007041'].plot()
```
